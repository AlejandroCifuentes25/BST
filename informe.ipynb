**Explicación de resultados:**

Los resultados fueron acordes con las tematicas vistas en la clase, dependiendo del arbol en cuestion y con su respectivo orden se pueden ver claras diferencias en las estructuras finales de estos, en donde las listas desorganizadas fueron de menor altura y con un coste promedio de O(log(n)), el cual es un coste barato y eficiente para la programación, sus alturas reducidas debido a que se encuentran con valores menores y mayores, los cuales favorecen a la ramificación con dos hijos o un grado maximo de 2, en cambio, las listas organizadas tienden a tener una mayor altura, menos hijos o con un grado maximo de 1, asi con un coste promedio de O(n), que comparado con las listas aleatorias, es de una indola cara.

Dejando asi la comparación clara del coste entre ambas listas.

**Análisis crítico:**

Se puede analizar que la eficiencia de un arbol binario de busqueda(BST) depende exclusivamente de la forma en la que se inserten los valores, donde las listas con valores aleatorios o balanceadas producen arboles que son mas eficientes y que en cambios listas con valores organizados requieren mayor eficiencia y por lo tanto generan un mayor costo y complejidad. Por el lado de los recorridos, estos ayudando ampliamente a saber la estructura interna de los arboles, los cuales son utiles para realizar operaciones varias, ademas de poder saber su profundidad, que referente a eso, la altura nos remarca la eficiencia de los arboles, donde con una altura menor genera mayor eficiencia y por el contrario, genera una menor eficiencia, concluyendo que la altura es directamente proporcional al coste en el peor caso. Para añadir, una de las facilidades de estos arboles con su invariante, es la busqueda de un valor, pues gracias a sus ramificaciones facilmente se puede ahorrar la mitad de un camino con el simple hecho de comparar si el valor es mayor o menor que la raiz principal.
